```python
import os

import matplotlib.pyplot as plt
```

```python
parent_folder = os.path.join('..', 'data_all', 'training')
counts = {}

for class_folder in os.listdir(parent_folder):
    class_path = os.path.join(parent_folder, class_folder)

    if os.path.isdir(class_path):
        num_files = len([f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))])
        counts[class_folder] = num_files

for class_folder, count in counts.items():
    print(f"The folder '{class_folder}' contains {count} files")
```

```python
folders = list(counts.keys())
counts = list(counts.values())

plt.bar(folders, counts)
plt.xlabel('Lesion')
plt.ylabel('Observations count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
